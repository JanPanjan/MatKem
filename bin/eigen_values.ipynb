```python
def eigs(g: nx.Graph) -> tuple[np.ndarray, np.ndarray]:
    """ Returns eigen values and eigen vectors of a graph. """
    A = nx.to_numpy_array(g)
    eigval, eigvec = np.linalg.eigh(A)
    eigval, eigvec = (np.round(eigval, 2), np.round(eigvec, 2))
    return (eigval, eigvec)
```

```python
eigval, _ = eigs(benzenoid_linear_2)
nc = {}
eigval
```

```python
for val in eigval:
    if val not in nc:
        nc[val] = 1
    else:
        nc[val] += 1

nc
```

```python
ncc = list(nc.values())
ncc
```

```python
v = [0] * len(eigval)
cur_id = 0
num_orb = len(eigval)
print("num orb:", num_orb)

for n in ncc:
    r = range(cur_id, cur_id + n)
    fill = 2 * n
    print("n:", n)
    print("range:", r)
    print("cur_id:", cur_id)
    while fill > 0 and num_orb > 0:
        b = False
        for _ in range(n):
            for i in r:
                v[i] += 1
                fill -= 1
                num_orb -= 1
                if num_orb == 0:
                    b = True
                    break
            if b:
                break
            print(" v:", v, end=",")
            print(" fill:", fill)
    cur_id += n
```
