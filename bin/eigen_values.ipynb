```python
import numpy as np
import networkx as nx
from pprint import pprint
```

```python
g = nx.Graph([
    (1,2), (1,5), (1,8),
    (2,3), (2,6),
    (3,4),
    (4,5),
    (6,7),
    (7,8)
])
A = nx.to_numpy_array(g)
A
```

```python
eigval, eigvec = np.linalg.eigh(A)
eigval
```
